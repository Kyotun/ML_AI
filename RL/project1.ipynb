```python
environment_name = 'ALE/Breakout-v5'
environment = gym.make(environment_name)
environment.reset()
```

```python
environment.action_space
```

```python
environment.observation_space
```

```python
episode = 5
for episode in range (1, episode+1):
    obs = environment.reset()
    done = False
    score = 0
    
    while not done:
        #environment.render()
        action = environment.action_space.sample()
        obs, reward, done, info, extra_info = environment.step(action)
        score += reward
    print('Episode:{}, Socre:{}', episode, score)
```

```python
environment.close()
```

### Vectorise the Environment

```python
environment = make_atari_env(environment_name, n_envs=4, seed=0)
environment = VecFrameStack(environment, n_stack=4) 
```

```python
environment.reset()
```

```python
environment.render()
```

```python
log_path = os.path.join('Training', 'Logs')
model = A2C('CnnPolicy', environment, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=100000)
```

### Save and Reload the Model

```python
a2c_path = os.path.join('Training', 'Saved Models', 'A2C_Breakout_Model')
model.save(a2c_path)
```

```python
del model
```

```python
model = A2C.load(a2c_path, environment)
```

### Evaluate and Test

```python
#Note: We can evaluate only in 1 environment, we created 4 for training
environment = make_atari_env(environment_name, n_envs=1, seed=0)
environment = VecFrameStack(environment, n_stack=4)
```

```python
evaluate_policy(model, environment, n_eval_episodes=10, render=True)
```
