```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()

from scipy import stats
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
```

```python
raw_data = pd.read_csv('../Part_5_ML/S36_L243/binary_predictors.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
data['Gender'] = data['Gender'].map({'Female':1, 'Male':0})
data['Admitted'] = data['Admitted'].map({'Yes':1, 'No':0})
data
#Male and No are references
```

```python
data.describe()
```

```python
y = data['Admitted']
x1 = data['Gender']
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
results_log.summary() # our model
#p-value is 0.000, significant
#Gender is significant
#log(odds) = -0.64 + 2.08*Gender
#log(ODDSfemale/ODDSmale) = 2.08*(1-0) , 1 for Female 0 for Male
#ODDSfemale = 7.99*ODDSmale
#Female has 8 more chance to admittance compare a man
```

```python
np.exp(2.0786)#coef of gender
```

```python
y = data['Admitted']
x1 = data[['SAT','Gender']]
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
results_log.summary()
#Log-likelihood is increased, LLR p-value is decreased
#Therefore model is more significant
#P-value of Gender has changed to 0.022 rather than 0.000, cause of SAT
#Because, SAT is a outstanding predictor
```

```python
np.exp(1.9449) #coef of new gender after SAT is given
#Female has 7 times more chance by admittance when Male and Famale both have same SAT score

#The more the data such SAT, Gender, Skills, etc. the better the estimation and interpretation
```

```python
test_actual = test['Admitted'] #actual outcomes that were observed
test_data = test.drop(['Admitted'], axis=1) # drop the outputs
test_data = sm.add_constant(test_data) # add constant and voila! here is our input variable-
#test_data = test_data[x.columns.values]
test_data
```

```python
def confusion_matrix(data, actual_values, model): 
    pred_values = model.predict(data) #predict the values using logit model
    bins = np.array([0,0.5,1]) # create a histogram that 0-0.5 round to 0, 0.5-1 round to 1
    cm = np.histogram2d(actual_values, pred_values, bins=bins)[0] #confusion matrix
    accuracy = (cm[0,0]+cm[1,1])/cm.sum() # calculate accuracy
    return cm, accuracy

#data: data frame or array
    #is formatted same way as input without actual(observed) values
    # const, var1, var2, ..., order is important
#actual_values: data frame or array
    #actual(observed) values from test_data
    #in logistic regression this should be single column with 0s and 1s
    
```

```python
cm = confusion_matrix(test_data, test_actual, results_log) # cm : cm and accuracy
cm
#training accuracy is higher than test accuracy cause of overfitting
```

```python
cm_df = pd.DataFrame(cm[0])
cm_df.columns = ['Predicted 0', 'Predicted 1']
cm_df = cm_df.rename(index={0:'Actual 0', 1:'Actual 1'})
cm_df
```
