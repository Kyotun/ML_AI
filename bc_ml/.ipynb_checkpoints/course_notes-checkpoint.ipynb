```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("melbourne.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df_copy = df.copy()
numerical_columns = df_copy.select_dtypes(include=['float64','int64'])
numerical_columns
```

```python
numerical_columns["Price"] = numerical_columns["Price"] / 10**3
df_rounded_price = df_copy
df_rounded_price[numerical_columns.columns] = numerical_columns.round(1)
df_rounded_price
```

```python
df_rounded_price
```

```python
plot_df = df_rounded_price[["Rooms", "Price"]]
plot_df.plot(x="Rooms", y="Price", xlabel="Number of Rooms", ylabel="Prices in K", kind="scatter")
```

```python
df_one_room = df_rounded_price.loc[df_rounded_price["Rooms"] == 1]
df_one_room["Price"].mean()
```

```python
df_scaled_dropped = df_rounded_price.dropna()
df_scaled_dropped = df_scaled_dropped.reset_index(drop=True)
df_scaled_dropped.groupby("Rooms")["Price"].mean()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_obj = LinearRegression()
```

```python
x = df_scaled_dropped["Rooms"].values.reshape(-1,1)
y = df_scaled_dropped["Price"].values.reshape(-1,1)
reg_obj.fit(x,y)
```

```python
reg_obj.coef_
#slope
```

```python
reg_obj.intercept_
#
```

```python
test = [4,5,8,30]
df_test = pd.DataFrame(test, columns=["Rooms"])
reg_obj.predict(df_test.values.reshape(-1,1))
```

```python
reg_obj.score(x,y)
```
